```python
import random
from random import randint
#function helps to validate  whether input contains only alphabets#
def get_name():
    user_name = input()
    while not user_name.isalpha():
        print("Don't be shy!What is your name?\n")
        user_name = input()
    print("Great",str(user_name),",Let's start!What is your gender? Enter M for male or F for female.\n")
    return user_name
#function helps to validate  whether input contains only specfic characters#
def get_gender():
    user_gender=input()
    gender=['M','F','m','f']
    while user_gender not in gender:
        print("I’m sorry, I cannot understand. What is your gender? Enter M for male or F for female.")
        user_gender=input()
    return user_gender
#function helps to validate  whether input contains only numbers and input value ranges in the particular limit#
def get_number(min,max,prompt,error):
    number=input(prompt)
    while not number.isdigit() or (int(number)<min or int(number)>max):
        print(error)
        number = input(prompt)
    return int(number)
#function helps to calculate the rdci value based on the user input related to weight,height,age#
def calculate_rdci(weight,height,age,gender):
    rdci=10*weight+6.25*height-5*age
    if gender.upper()=='M':
        calories_per_day=rdci+5
    else:
        calories_per_day=rdci-161
    print("Considering the details given, your daily recommended intake is of", calories_per_day, "calories per day.")
    return rdci

#function to calculate the daily calories#
def calculate_calories(rdci,options,error):
    total_calories = []
    calorie_count = 0
    for day in range(1, 8):
        print("Day" + str(
            day) + ": were your meals very unhealthy (1), unhealthy (2), healthy (3), or very healthy (4)?\n Enter the corresponding number:\n")
        daily_intake = input()
        while not daily_intake.isdigit() or daily_intake not in options:
            print(error)
            daily_intake = input()
        if daily_intake == '1':
            calorie_count = rdci*1.5
        elif daily_intake == '2':
            calorie_count = rdci*1.2
        elif daily_intake == '3':
            calorie_count = rdci
        else:
            calorie_count = rdci*0.8
        total_calories.append(calorie_count)
    return total_calories
#function helps to calculate radom temptations towards food#
def calculate_temptation(calorie_value):

    temptation_dict={'chocolate':250,'chips':550,'ice-cream':207,'fast-food':350,'fizzy drink':180,'party cake':257,'popcorn':375}
    random_temptation=random.choice(list(temptation_dict.values()))
    tempted_key=list(temptation_dict.keys())[list(temptation_dict.values()).index(random_temptation)]
    random_number = randint(0, 1)
    if random_number==1:
         print("Also, it looks like this day you’ve been tempted with", tempted_key, "and", random_temptation,"calories have been added!")
         return calorie_value+random_temptation
    else:
        return calorie_value
#function to calculate adci#
def calculate_adci(calories):
    temptation_calories=[]# to store calories after temptation is added#
    for i in range(len(calories)):
        print("{:.2f}".format(calories[i]), "calories in take in day",i+1)
        temptation_calories.append(calculate_temptation(calories[i]))
    average_calorie_value=sum(temptation_calories)/len(calories)
    print("During the last 7 days you had an intake of","{:.2f}".format(sum(temptation_calories)),"calories, meaning a daily average of","{:.2f}".format(average_calorie_value),"calories.")

    return average_calorie_value
#functions gives the recommendations based on the adci and rdci values#
def calorie_recommendation(username,rdci,adci):
    percentage_value=(adci/rdci)*100
    if percentage_value<90:
        print(username,"your daily calorie intake is lower than the recommended with {:.2f}".format(percentage_value-100),"%. This way you will lose weight, just make sure your meals contain all nutritional value needed. It’s recommended that you do not fall under the healthy weight and that you keep a balanced lifestyle")
    elif percentage_value>90 and percentage_value<110:
        print(username,"your daily calorie intake is close to the recommended one! You have a balanced healthy lifestyle, well done!")
    else:
        print(username,",your daily calorie intake is higher than the recommended with {:.2f}".format(percentage_value-100),"%. This way you will gain weight, and in time health concerns may arise. It’s recommended that you either lower your calorie intake, either exercise more!")
    return percentage_value
def main():
    age_prompt=("What is your age in years?\n")
    age_error=("I’m sorry, I cannot understand. What is your age in years?\n")
    weight_prompt=("What is your current weight in kg?\n")
    weight_error=("I’m sorry, I cannot understand. What is your weight in kg?\n")
    height_prompt=("What is your current height in cm?\n")
    height_error=("I’m sorry, I cannot understand. What is your height in cm?\n")
    calorie_intake_error=("I’m sorry, I cannot understand.Were your meals very unhealthy (1), unhealthy (2), healthy (3), or very healthy (4)? Enter the corresponding number")

    options=['1','2','3','4']
    print("Welcome to the Calorie Intake Assistant,my name is cal,what is your name?\n")
    user_name=get_name()
    user_gender=get_gender()
    user_age=get_number(18,75,age_prompt,age_error)
    user_weight=get_number(40,150,weight_prompt,weight_error)
    user_height=get_number(100,200,height_prompt,height_error)
    print("Thank you for the information",str(user_name),"!")
    rdci=calculate_rdci(user_weight,user_height,user_age,user_gender)
    print("Let's see how healthy your meals were last week")
    calories_intake=calculate_calories(rdci,options,calorie_intake_error)
    print(str(user_name), ",here are your results")
    adci=calculate_adci(calories_intake)
    calorie_recommendation(user_name,rdci,adci)
    print("Goodbye and good luck!")

main()
```
